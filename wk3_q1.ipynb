```python
def swap1(x,y) :
    x=y
    y=x
    return(x,y)

def swap2(x,y) :
    return(y,x)

def swap3(x,y) :
    z=x
    x=y
    y=z
    return(x,y)

def swap4(x,y) :
    x,y=y,x
    return(x,y)
```

```python
x=1
y=2
(x,y) = swap1(x,y)
print("x=%s , y=%s" % (x, y))
```

```python
import random
def create_dna(n, alphabet='acgt'):
    return ''.join([random.choice(alphabet) for i in range(n)])
dna = create_dna(1000000)
```

```python
len(dna)
```

```python
dna
```

```python
def count1(dna, base):
    i = 0
    for c in dna:
        if c == base:
            i += 1 
    return i

def count2(dna, base):
    i = 0 
    for j in range(len(dna)):
        if dna[j] == base:
            i += 1 
    return i 

def count3(dna, base):
    match = [c == base for c in dna]
    return sum(match)

def count4(dna, base):
    return dna.count(base)

def count5(dna, base):
    return len([i for i in range(len(dna)) if dna[i] == base])

def count6(dna,base):
    return sum(c == base for c in dna)
```

```python
t0 = time.time()
count1(dna, 'g')
print (time.time() - t0, "seconds process time")

```
